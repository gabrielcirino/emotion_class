```python
import matplotlib.pyplot as plt
```

```python
data = {
"emotions": {
"sadness": 0.4969351851851852,
"joy": 0.24302469135802468,
"fear": 0.42632407407407408,
"disgust": 0.53148148148148144,
"anger": 0.383900462962963
},
"sentiment": {
"score": -0.18433796296296295,
"label": "NEGATIVE"
}
}
```

```python
emotions = data['emotions']
labels = emotions.keys()
sizes = emotions.values()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribuição de Emoções')
plt.show()
```
